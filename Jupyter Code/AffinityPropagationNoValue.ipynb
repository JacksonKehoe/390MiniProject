```python
import numpy  as np
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.cluster import AffinityPropagation
```

```python
df = pd.read_csv("../Processed Data/Finance_only_overunder.csv")
# convert 'Ticker' column into row indices
df.set_index('Ticker', inplace=True)

# remove sector, company and number column
df_sector = df['Sector']
df = df.drop(columns=['Sector'])
df_company = df['Company']
df = df.drop(columns=['Company'])
df_number = df['Number']
df = df.drop(columns=['Number'])
df_overunder = df['over/under']
df = df.drop(columns=['over/under'])
#normalizing the data
normalized_df = (df-df.mean())/df.std()

# display
display(normalized_df)
```

```python
random_state = 0
af = AffinityPropagation(preference=-100).fit(normalized_df)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
labels_pred = af.predict(normalized_df)
```

```python
n_clusters_ = len(cluster_centers_indices)
print('Estimated number of clusters: %d' % n_clusters_)
```

```python
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(df, labels, metric='sqeuclidean'))
```
